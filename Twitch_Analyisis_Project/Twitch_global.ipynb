```python
year_list=list(data.year.unique())
year_list
```

```python
average_streams=[]

for i in year_list:
    x=data[data.year==i]
    year_average=sum(x.Streams)/len(x)
    average_streams.append(year_average)

average_streams
```

```python
df1=pd.DataFrame({'year_list':year_list,'average_streams':average_streams})
df1.head(6)
```

```python
plt.figure(figsize=(15,10))
plt.bar(year_list,average_streams)

plt.xticks(rotation=0)
plt.xlabel('Year')
plt.ylabel('Average number of streams')
plt.title('Average number of streams per year')
```
