# Labtop Data Investigation

In this notebook, I'll be looking into the a data set named "Laptop Specs and Latest price." This data set can be found in Kaggle with this link: https://www.kaggle.com/datasets/kuchhbhi/latest-laptop-price-list

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
os.getcwd()
```

```python
laptop=pd.read_csv('Cleaned_Laptop_data.csv')
laptop.head(10)
```

In the columns named latest_price and old_price the numbers are based in Indian Rubies. I want these values to be based in American Dollars so I search up the conversion to rubies to U.S dollars. That value was 73.9339 so I divided the entire column values by that value. 

```python
laptop['latest_price']=laptop['latest_price']/73.9339
laptop['old_price']=laptop['old_price']/73.9339
laptop.head(10)
```

Let's see if the there is different brands of labtops in this data set:

```python
laptop.brand.unique()
brand=laptop['brand'].value_counts()
brand
```

Now let's normalise this set by using percentages. The brands specific counts divided by the total amount of counts the data set has. 

```python
bran=brand.rename_axis("Brand").reset_index(name='counts')
bran['percent'] = (bran['counts'] / bran['counts'].sum()) * 100
bran
```

```python
percent=bran["percent"]
brand=bran["Brand"]
```

Let's visualized the percentages by using a pie chart

```python
f, ax = plt.subplots(1,1, figsize= (10,10))
ax.pie(percent,labels= brand,startangle= 180,autopct = '%1.2f%%')
ax.legend()
```

We can see there are more brands from ASUS, DELL, LENOVO, and HP. These four make more than 75% of the data set so we will limit the set to these four.

```python
sm_laptop=laptop[(laptop.brand=='ASUS')|
                 (laptop.brand=='DELL')|
                 (laptop.brand=='Lenovo')|
                 (laptop.brand=='HP')]
sm_laptop=sm_laptop.reset_index()
sm_laptop
```

```python
sm_laptop.dtypes
```

```python
sm_laptop['brand'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
column=list(sm_laptop.columns)
column
```

```python
scaler = OrdinalEncoder()
names = sm_laptop.columns
d = scaler.fit_transform(sm_laptop)

scaled_df = pd.DataFrame(d, columns=names)
scaled_df.head(10)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, GridSearchCV
import xgboost as xgb

from sklearn.metrics import precision_score,recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix,accuracy_score, classification_report
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
features, labels=  oversample.fit_resample(scaled_df.drop(["brand"],axis=1),scaled_df["brand"])
```

```python
X_train, X_test, y_train, y_test=train_test_split(features, labels,test_size=0.33,random_state=42)
```

```python
models = [RandomForestClassifier(), KNeighborsClassifier(), SVC(), LogisticRegression(),xgb.XGBClassifier()]
scores = dict()

for m in models:
    m.fit(X_train, y_train)
    y_pred = m.predict(X_test)

    print(f'model: {str(m)}')
    print(classification_report(y_test,y_pred, zero_division=1))
    print('-'*30, '\n')
```

```python
model=RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test,y_pred, zero_division=1))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(model.predict(X_test),y_test)
disp = ConfusionMatrixDisplay(cm, display_labels=["Lenovo","HP","ASUS","DELL"])
disp.plot()
plt.title("Confusion Matrix using random forest")
plt.show()
```

```python
model=xgb.XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test,y_pred, zero_division=1))
```

```python
cm = confusion_matrix(model.predict(X_test),y_test)
disp = ConfusionMatrixDisplay(cm, display_labels=["Lenovo","HP","ASUS","DELL"])
disp.plot()
plt.title("Confusion Matrix")
plt.show()
```
